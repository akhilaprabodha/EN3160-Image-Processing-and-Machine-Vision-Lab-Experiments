# Image Creation

#### Creating a (6, 8) grayscale image with a white pixel at (2, 3)

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im = np.zeros((6,8), dtype=np.uint8) # 2D means Gray scale image
im[2,3] = 255

fig, ax = plt.subplots(1, 1, figsize=(6, 8))
ax.imshow(im, cmap='gray', vmin=0, vmax=255)
ax.xaxis.set_ticks_position('top')
plt.show()
```

#### Creating a (6, 8) color image with a pink pixel at (2, 3)


```python
im = np.zeros((6,8,3), dtype=np.uint8) # 3D means colour image with RGB 3 channels
im[2,3] = (255, 190, 203)

fig, ax = plt.subplots(1, 1, figsize=(6, 8))
ax.imshow(im)
ax.xaxis.set_ticks_position('top')
plt.show()
```

#### Opening an Image and Displaying Using Matplotlib.


```python
im = cv.imread('images/tom.jpg')
assert im is not None

fig, ax = plt.subplots(1, 1, figsize=(6, 8))
ax.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax.xaxis.set_ticks_position('top')
plt.show()
```

#### Opening an Image and Displaying Using OpenCV.

```python
im = cv.imread('images/tom.jpg')
assert im is not None

cv.namedWindow('Image', cv.WINDOW_AUTOSIZE) 
cv.imshow('Image', im)
cv.waitKey(0) # press any key to exit
cv.destroyAllWindows()
```

#### Opening an Image and Displaying Properties.


```python
im = cv.imread('images/tiger.jpg')
assert im is not None

fig, ax = plt.subplots(1, 1, figsize=(6, 8))
ax.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax.xaxis.set_ticks_position('top')
plt.show()

print('Image Shape:', im.shape)
print('Image Data Type:', im.dtype)
print('Image Size:', im.size)
```

#### Open image axis-off

```python
im = cv.imread('images/emma.jpg')
assert im is not None

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB)) # convert to RGB because matplotlib likes RGB
# ax.imshow(im) # wrong
ax.axis('off')
plt.show()
```

#### Inceace image brightness [INCORRECT]

```python
im1 = cv.imread('images/emma.jpg', cv.IMREAD_GRAYSCALE)
assert im1 is not None

im2 = im1 + 100 # increase brightness

fig, ax = plt.subplots(1, 2,figsize=(10, 5))
ax[0].imshow(im1, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original')
ax[0].axis('off')
ax[1].imshow(im2, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Enhanced brightness')
ax[1].axis('off')
plt.show()
```

#### Inceace image brightness [CORRECT METHOD]

```python
im2 = cv.add(im1, 100) # increase brightness

fig, ax = plt.subplots(1, 2,figsize=(10, 5))
ax[0].imshow(im1, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original')
ax[0].axis('off')
ax[1].imshow(im2, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Enhanced brightness')
ax[1].axis('off')
plt.show()
```

#### Extracting the `Blue` Channel from an Image.

```python
im = cv.imread('images/fantasy-marine-landscape-with-bioluminescent-nature.jpg')
assert im is not None

im_blue = im.copy()
im_blue[:,:,1] = 0 # Green colour to zero
im_blue[:,:,2] = 0 # Red colour to zero

fig, ax = plt.subplots(1, 2,figsize=(10, 5))
ax[0].imshow(im, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original')
ax[0].axis('off')
ax[1].imshow(im_blue, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Blue channel')
ax[1].axis('off')
plt.show()
```

#### Extracting `Red` and `Green` channels

```python
im_red = im.copy()
im_red[:,:,0] = 0 # Blue colour to zero
im_red[:,:,1] = 0 # Green colour to zero

im_green = im.copy()
im_green[:,:,0] = 0 # Blue colour to zero
im_green[:,:,2] = 0 # Red colour to zero

fig, ax = plt.subplots(1, 2,figsize=(10, 5))
ax[0].imshow(im_red, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Blue channel')
ax[0].axis('off')
ax[1].imshow(im_green, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Green channel')
ax[1].axis('off')
plt.show()
```

#### Image Conversion to HSV Color Space

```python
im = cv.imread('images/pencils.jpg')
assert im is not None

im_hsv = cv.cvtColor(im, cv.COLOR_BGR2HSV)

fig, ax = plt.subplots(2, 2, figsize=(12, 8))
ax[0, 0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax[0, 0].set_title('Original Image')
ax[0, 1].imshow(im_hsv[:,:,0], cmap='gray')
ax[0, 1].set_title('Hue Channel')
ax[1, 0].imshow(im_hsv[:,:,1], cmap='gray')
ax[1, 0].set_title('Saturation Channel')
ax[1, 1].imshow(im_hsv[:,:,2], cmap='gray')
ax[1, 1].set_title('Value Channel')
for a in ax.ravel():
    a.axis('off')
plt.show()
```
